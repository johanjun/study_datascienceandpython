```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 점선그래프

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

### 다중 그래프

```python
plt.subplot(1, 2, 1)
plt.subplot(1, 2, 2)
plt.show()
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2, 2, 1)
plt.plot(data)

plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)

plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30) + 3 + np.random.randn(30))

plt.show()
```

### option

```python
plt.plot(data, 'k+')
plt.show()
```

```python
plt.plot(data, 'g--')
plt.show()
```

### size

```python
plt.figure(figsize=(10,5))
plt.plot(data, 'k+')
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(2, 2, 1)
plt.plot(data)

plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)

plt.subplot(2, 2, 3)
plt.scatter(scat_data, np.arange(30) + 3 + np.random.randn(30))

plt.show()
```

### 그래프 겹치기 + legend

```python
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend()
plt.show()
```

### 이름 달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title('random graph')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.show()
```

### 종합

```python
plt.figure(figsize=(10,5))
plt.plot(np.random.randn(1000).cumsum(),
        'k', label='one')
plt.plot(np.random.randn(1000).cumsum(),
        'b', label='two')
plt.plot(np.random.randn(1000).cumsum(),
        'r', label='three')
plt.legend()
plt.show()

#save
plt.savefig('saved_graph.svg')
```
