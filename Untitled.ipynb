```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = './1.jpg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
image.shape
```

### 보기

```python
image.shape
```

```python
np.min(image), np.max(image)
```

```python
plt.hist(image.ravel(), 255, [0, 256])
plt.show()
```

### 그림 나타내기

```python
plt.imshow(image)
plt.show()
```

### 흑백으로

```python
image_pil = Image.open(path).convert('L')
image_bw = np.array(image_pil)
image_bw.shape
```

```python
plt.imshow(image_bw, 'gray')
plt.show()
```

```python
#높을수록 r, 낮을수록 b
plt.imshow(image_bw, 'jet')
plt.show()
```

### colorbar

```python
plt.imshow(image_bw, 'jet')
plt.colorbar()
plt.show()
```

### 이미지 설정

```python
plt.figure(figsize=(10,10))
plt.imshow(image)
plt.show()
```
