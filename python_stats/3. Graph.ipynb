```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.arange(-1.0, 1.0, 100)
```

# 1

```python
#(a)
x = np.linspace(-1, 1, 100)
y = x**2
```

```python
#(b, c)
plt.scatter(x, y, color='r')
```

# 2

```python
#(a)
x = np.linspace(0, 6)
y1 = x
y2 = x**2
y3 = np.log(x)
y4 = np.sqrt(x)
```

```python
plt.subplot(221)
plt.scatter(x, y1)
plt.subplot(222)
plt.scatter(x, y2)
plt.subplot(223)
plt.scatter(x, y3)
plt.subplot(224)
plt.scatter(x, y4)
plt.show()
```

```python
#(b)
# plt.subplot(221)
plt.plot(x, y1)
# plt.subplot(222)
plt.plot(x, y2)
# plt.subplot(223)
plt.plot(x, y3)
# plt.subplot(224)
plt.plot(x, y4)
plt.show()
```

```python
#(c)
plt.scatter(x, y1)
plt.plot(x, y2)
plt.plot(x, y3, linestyle='-.')
plt.plot(x, y4, linestyle='-.')
plt.show()
```

# 3

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.ylim(-5, 5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```
